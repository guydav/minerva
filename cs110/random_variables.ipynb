```python
from numpy import random, average, log
from math import factorial
import matplotlib.pyplot as plt


def hire_assistant(n):
    applicant_pool = random.uniform(size=n)
    current_assistant = 0
    hirings = 0

    for assistant in applicant_pool:
        if assistant > current_assistant:
            current_assistant = assistant
            hirings += 1

    return hirings


def hire_assistant_min_threshold(n, threshold=0.1):
    max_hire_possible = 1 - threshold
    applicant_pool = random.uniform(size=n)
    current_assistant = 0
    hirings = 0

    for assistant in applicant_pool:
        if assistant > current_assistant + threshold:
            current_assistant = assistant
            hirings += 1
            if current_assistant >= max_hire_possible:
                break

    return hirings


def optimized_hiring_decision(n, daily_work, employment_agencey_fee, retrenchment_fee):
    applicant_pool = random.uniform(size=n)
    current_assistant = 0
    hirings = 0

    for index in xrange(n):
        assistant = applicant_pool[index]
        if (assistant > current_assistant) and \
            (daily_work * (assistant - current_assistant) > employment_agencey_fee + retrenchment_fee):
                current_assistant = assistant
                hirings += 1

    return hirings


def simulate_hiring(max_n=100000, trials=10):
    n_values = range(2, max_n)
    results = []

    for n in n_values:
        results.append(average([hire_assistant(n) for _ in xrange(trials)]))
    
    return n_values, results


def simulate_hiring_threshold(threshold=0.1, max_n=1000, trials=100):
    n_values = range(2, max_n)
    results = []

    for n in n_values:
        results.append(average([hire_assistant_min_threshold(n, threshold) for _ in xrange(trials)]))
    
    return n_values, results


def simulate_hiring_threshold_probability_of_x(threshold=0.1, x=2, max_n=1000, trials=100):
    n_values = range(2, max_n)
    results = []

    for n in n_values:
        results.append(average([hire_assistant_min_threshold(n, threshold) == x for _ in xrange(trials)]))
    
    return n_values, results


def plot_hiring_min_threshold():
    results = {t:simulate_hiring_threshold(t) for t in (0.01, 0.05, 0.1, 0.2)}

    plt.figure(figsize=(12, 8), dpi=400)
    
    for t in results:
        x, y = results[t]
        plt.plot(x, y, label='Simulated Results for threshold = {t}'.format(t=t))
    
    plt.grid(True)
    plt.title("Hiring Problem With Threshold, simulated results")
    plt.xlabel("Candidate pool size")
    plt.ylabel("Total candidate hired")
    plt.legend()
    
    plt.show()
    
    results = {t:simulate_hiring_threshold_probability_of_x(t, x=2) for t in (0.01, 0.05, 0.1, 0.2)}

    plt.figure(figsize=(12, 8), dpi=400)
    
    for t in results:
        x, y = results[t]
        plt.plot(x, y, label='Simulated odds of hiring exactly two for threshold = {t}'.format(t=t))
    
    plt.grid(True)
    plt.title("Hiring Problem With Threshold, simulated results")
    plt.xlabel("Candidate pool size")
    plt.ylabel("Total candidate hired")
    plt.legend()
    
    plt.show()


def plot_hiring_results(x, y):
    plt.figure(figsize=(12, 8), dpi=400)
    plt.plot(x, y, c='r', label='Simulated Results')
    plt.plot(x, log(x), c='b', label='Expected Results (ln n)')
    
    plt.grid(True)
    plt.title("Hiring Problem, expected and simulated results")
    plt.xlabel("Candidate pool size")
    plt.ylabel("Total candidate hired")
    plt.legend()
    
    plt.show()
    
        
def hat_check(n):
    hats = range(n)
    random.shuffle(hats)
    correct = [x == hats[x] for x in xrange(n)]
    return sum(correct)    


def simulate_hats(max_n=50, trials=100):
    n_values = range(10, max_n)
    results = []

    for n in n_values:
        results.append(average([hat_check(n) for _ in xrange(trials)]))

    return n_values, results


probability_function_cache = {}
probability_function_cache[0] = 1
max_n = 0
def probability_function(n):
    global probability_function_cache, max_n
    if n <= max_n:
        return probability_function_cache[n]
    
    for i in xrange(max_n + 1, n + 1):
        probability_function_cache[i] = probability_function_cache[i - 1] + ((-1.0) ** i / factorial(i))
            
    max_n = n
    return probability_function_cache[n]


def plot_hats_simulation(x, y):
    plt.figure(figsize=(12, 8), dpi=400)
    plt.plot(x, y, c='r', label='Simulated Results')
    plt.plot(x, [1] * len(x), c='b', label='Expected Results')
    
    plt.grid(True)
    plt.title("Hat check problem, expected and simulated results")
    plt.xlabel("Number of guests")
    plt.ylabel("Hats returned correctly")
    plt.legend()
    
    plt.show()

        
def main():
    #     plot_hiring_results(*simulate_hiring())
    #     plot_hats_simulation(*simulate_hats())
    plot_hiring_min_threshold()


if __name__ == '__main__':
    main()
```
