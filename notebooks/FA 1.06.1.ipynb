```python
x = np.linspace(-8, 8, 2000)
plt.figure(figsize = (8,8), dpi = 400)
a = -8
b = 0
c = -8


def generate_plot(x, a, b, c):
    half = np.sqrt(a * (x **3) + b * (x ** 2) + c * x)
    plt.plot(x, half, x, -1 * half)
    return 'a = {a}, b = {b}, c = {c}'.format(a=a, b=b, c=c)

def plot_range(x, a, b, c, min=0, max=10):
    legends = []
    
    for i in xrange(min, max):
        legends.append(generate_plot(x, a + i, b + i, c + i))
        
    plt.legend(legends)
    plt.title('a = {a}+i, b = {b}+i, c = {c}+i, i in range({min}, {max})'.format(a=a, b=b, c=c, min=min, max=max))
    
plot_range(x, a, b, c)
```

```python
x = np.linspace(0, 20, 2000)
plt.figure(figsize = (8,8), dpi = 400)

def sin_plot(x, a, b):
    plt.plot(x, a * np.sin(x * b))

sin_plot(x, 1, 1)
sin_plot(x, 1, 0.5)
sin_plot(x, 1, 2)
plt.title('{a} * sin(b * t)'.format(a=a, b=b))
```

```python
x = np.linspace(0, 20, 5000)
plt.figure(figsize = (8,8), dpi = 400)

def fourier_plot(x, a, w, a2, w2, shift):
    plt.plot(x, a * np.sin(w * x) + a2 * np.sin(w2 * x + shift))

fourier_plot(x, 1, 1, 0.5, 3.37, -0.5 * np.pi)
plt.title('Fourier')
```

```python
from matplotlib import patches
plt.figure(figsize = (8,6), dpi = 800)
total_range = np.linspace(900,1100,2000)
fifty_percent_range = np.linspace(900, 1013, 1130)

def poisson_range(x, l = 1000):
    return np.array([stats.poisson(l).cdf(n) for n in x])

plt.plot(total_range, poisson_range(total_range))
plt.stackplot(fifty_percent_range, poisson_range(fifty_percent_range), alpha=0.75, colors=['#74BBFB'], baseline='zero')
plt.xlabel('Number of mutations')
plt.ylabel('Cumulative Probability')
poisson_patch = patches.Patch(color='blue', linestyle='solid', label='Poisson CDF (lambda = 1000)')
fifty_percent_patch = patches.Patch(color='#74BBFB', label='CDF for P_parthenogensis = 0.5\nk = 1013, P_k = 0.667')
plt.legend(handles = [poisson_patch, fifty_percent_patch], loc = 'best')
```
