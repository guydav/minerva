```python
import pandas
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import networkx as nx
matplotlib.style.use('ggplot')
```

```python
graph_data = {'A': ['B', 'C', 'F', 'G'], 'B': ['C', 'E'], 'C': ['E'], 'D': ['E', 'I'],
             'E': ['G', 'I'], 'F': ['H', 'I'], 'G': ['H'], 'H': ['I']}

graph = nx.from_dict_of_lists(graph_data)
plt.figure(figsize=(12, 9))
nx.draw(graph, with_labels=True, node_color='c', alpha=0.8, node_size=500)
```

```python
nx.shortest_path(graph, 'A', 'D')
```

```python
spelling_data = {'moke': ['mode', 'mole', 'mike', 'mope', 'more'], 
                 'mode': ['made', 'code', 'mole', 'mope', 'more'],
                 'mole': ['mope', 'more', 'mile', 'mold'],
                 'more': ['mope', 'mire', 'mare'],
                }
spelling_graph = nx.from_dict_of_lists(spelling_data)
plt.figure(figsize=(12, 9))
nx.draw(spelling_graph, with_labels=True, node_color='#00FF55', alpha=0.60, node_size=1200)
```
