## 1. $\sqrt{N}$ Fluctuations

(c) Sketch a plot of what it would look like. 

```python
num_samples = 1000

for N in (100, 1000, 1000):
    samples = np.random.binomial(N, 0.5, num_samples)
    plt.hist(samples, density=True, bins=30, label='Samples')

    dist = stats.norm(N / 2, N ** 0.5 / 2)
    x_min = min(samples)
    x_max = max(samples)
    r = x_max - x_min
    x_min -= 0.1 * r
    x_max += 0.1 * r
    x_range = np.linspace(x_min, x_max)

    plt.plot(x_range, dist.pdf(x_range), label='Distribution')
    plt.title(f'N = {N}')
    plt.legend(loc='best')
    plt.show()
```
