```r
library(ggplot2)
library(geoR)
```

```r
nba.df <- read.csv('./nba_stats_fixed.csv')
nba.df$FTP <- nba.df$FT. / 100
nrow(nba.df)
```

```r
ggplot(nba.df, aes(x=FTP)) +
    geom_histogram(aes(y=..density.. ), binwidth=0.01) + 
    stat_function(fun=dnorm,args=list(mean=0.8, sd=0.08)) +
    theme(aspect.ratio=3/4) + 
    xlab("X") + 
    ylab("P(X)")
```
