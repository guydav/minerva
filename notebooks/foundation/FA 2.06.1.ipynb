```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

x = []
y = []
z = []

min_x_y = 2
max_x_y = 101
x_y_step = 5

def cohens_d(x1, s1, n1, x2, s2, n2):
    numerator = float(x1) - float(x2)
    denominator = (((n1 - 1) * s1 ** 2 + (n2 - 1) * s2 **2) / float(n1 + n2 - 2)) ** 0.5
    return numerator / denominator

for x_value in range(min_x_y, max_x_y, x_y_step):
    for y_value in range(min_x_y, max_x_y, x_y_step):
        x.append(x_value)
        y.append(y_value)
        z.append(cohens_d(155.0, 7.0, float(x_value), 148.0, 14.0, float(y_value)))

fig = plt.figure(figsize = (16,12), dpi = 800)
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z)

ax.set_xlabel('n1')
ax.set_ylabel('n2')
ax.set_zlabel("cohen's d")
```
