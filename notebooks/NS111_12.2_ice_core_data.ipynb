```python
import pandas
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')
```

```python
data_df = pandas.DataFrame.from_csv('icecoredata.csv')
```

```python
ax = data_df.plot(x='age_GISP', y='d18O_GISP', figsize=(16, 12))
data_df.plot(x='age_GRIP', y='d18O_GRIP', ax=ax, alpha=0.50)
ax.set_xlabel('Age (years)', fontsize=20)
ax.set_ylabel(r'$\delta ^{18}O$', fontsize=20)
```

```python
from scipy.interpolate import spline

ax = data_df.plot(x='age_GISP', y='d18O_GISP', alpha=0.50, figsize=(16, 12))
data_df.plot(x='age_GRIP', y='d18O_GRIP', ax=ax, alpha=0.50)
ax.plot(data_df['age_GISP'][:645][::3], 
        spline(data_df['age_GISP'][:645], data_df['d18O_GISP'][:645],  data_df['age_GISP'][:645][::3]),
       'red', linewidth=2.0, label='Smoothened d18O in GISP')
grip_ages = pandas.concat([data_df['age_GRIP'][:1000:7], data_df['age_GRIP'][1000::5]])
ax.plot(grip_ages, 
        spline(data_df['age_GRIP'], data_df['d18O_GRIP'], grip_ages),
       'steelblue', linewidth=2.0, label='Smoothened d18O in GRIP')
ax.set_xlabel('Age (years)', fontsize=20)
ax.set_ylabel(r'$\delta ^{18}O$', fontsize=20)
ax.invert_xaxis()
ax.legend(loc='best')
```
