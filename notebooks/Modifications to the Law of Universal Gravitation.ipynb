```python
import pandas
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')
```

```python
def first_r_solution(theta):
    """
    This is the solution for (M_p^2 \lambda)/l^2 = 0.1
    """
    omega = 3.0 / (10.0 ** 0.5)
    u = (-82.0 / 9) * np.cos(omega * theta) + (10.0 ** -5.5) / 3 * np.sin(omega * theta) + 100.0 / 9
    return 1.0 / u


def second_r_solution(theta):
    """
    This is the solution for the degenerate case, (M_p^2 \lambda)/l^2 = 1
    """
    u = 5.0 * theta ** 2 + (10 ** -6) * theta + 2
    return 1.0 / u
```

```python
MAX_PI = 20
VALUES_PER_PI = 100
theta_values = np.linspace(0, np.pi * MAX_PI, VALUES_PER_PI * MAX_PI)
fig = plt.figure(figsize = (16,12), dpi = 800)
ax = fig.add_subplot(111, projection='polar')

first_solution_values = first_r_solution(theta_values)
ax.plot(theta_values, first_solution_values, 
        label=r'First solution, $\frac{M_p^2 \lambda}{l^2} = 0.1$')

second_solution_values = second_r_solution(theta_values)
ax.plot(theta_values, second_solution_values, color='blue', 
        label=r'Second (degenerate) solution, $\frac{M_p^2 \lambda}{l^2} = 1$')

ax.set_rmax(0.505)
ax.grid(True)
ax.legend(loc=8)

x_labels = ['0', r'$\frac{\pi}{4}$', r'$\frac{\pi}{2}$', r'$\frac{3\pi}{4}$',
            r'$\pi$', r'$\frac{5\pi}{4}$', r'$\frac{3\pi}{2}$', r'$\frac{7\pi}{4}$']
ax.xaxis.set_ticklabels(x_labels)
ax.xaxis.set_tick_params(labelsize=24)
ax.yaxis.set_tick_params(labelsize=16)

ARROW_STEP = 4
for i in xrange(MAX_PI * 2 - 1):
    first_index = i * VALUES_PER_PI / 2 + 30
    plt.arrow(theta_values[first_index], first_solution_values[first_index], 
          theta_values[first_index + ARROW_STEP] - theta_values[first_index],
          first_solution_values[first_index + ARROW_STEP] - first_solution_values[first_index], 
          shape='full', lw=0, length_includes_head=True, head_width=.02, color='red', alpha=0.75)
    
    second_index = i * 6
    plt.arrow(theta_values[second_index], second_solution_values[second_index], 
              theta_values[second_index + ARROW_STEP] - theta_values[second_index],
              second_solution_values[second_index + ARROW_STEP] - second_solution_values[second_index], 
              shape='full', lw=0, length_includes_head=True, head_width=.02, color='blue', alpha=0.75)
```
