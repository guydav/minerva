```python
def f(N, x):
    return (1 + x / N) ** N
```

```python
N_values = [10 ** i for i in range(11)]
X_values = np.linspace(0, 10, 200)


plt.figure(figsize=(12, 9))

for N in N_values:
    plt.semilogy(X_values, f(N, X_values), label=r'$(1 + \frac{{x}}{{N}})^N$ for $N = {N}$'.format(N=N), alpha=0.75)

plt.plot(X_values, np.exp(X_values), label='$e^x$')

plt.legend(loc='best', fontsize=12)
plt.xlabel(r'x', size=12)
plt.ylabel(r'$\ln f(x)$', size=12)
plt.show()
```
